#### Customer subscription plan trend

- This bar line graph is to visualise the **change in subscription mode** over time for each customer.
- A sample of 8 customers is taken to have their subscription patterns analysed.
- The higher the plot is, the more advanced the subscription mode.
- The lowest, churn, means the customer has cancelled subscription.

```python
plan_order = ["churn", "trial", "basic monthly", "pro monthly", "pro annual"]
match_plan_id = {name: i for (i, name) in enumerate(plan_order)}

# remap values for plans
# churn will be mapped to 0, while the other plans will be mapped to their plan_id + 1
df = pd.read_csv('csv-files/customer_sample.csv')
df['plan_id'] = df['plan_name'].map(match_plan_id)

# plot
plt.figure(figsize=(10,5))
for customer_id in df['customer_id'].unique():
	customer = df[df['customer_id'] == customer_id]
	plt.plot(customer['start_date'], customer['plan_id'], marker='o', label=f'{customer_id}')

# customisation
plt.title('Customer Subscription Pattern', fontsize=14, pad=20)
plt.xlabel('Start Date', fontsize=12)
plt.ylabel('Plan Name', fontsize=12)
plt.legend('Customer ID')
plt.xticks(rotation=45)  # rotate x-axis labels -- dates are not cramped
plt.yticks(ticks=range(len(plan_order)), labels=plan_order) # switch the ticks to 1, 2, 3, 4, ... then replace the numbers with str in place_order
plt.show()
```
